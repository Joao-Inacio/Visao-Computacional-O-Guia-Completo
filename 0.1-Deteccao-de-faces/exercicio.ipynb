# **Exercício de Detecção**

```python
import cv2
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
BASE_DIR = Path.cwd().parent

caminho_imagem = BASE_DIR / "data" / "Images"
classifier = BASE_DIR / "data" / "Cascades"
```

## Detecção de carros

```python
imagem_carro = cv2.imread(str(f"{caminho_imagem}/car.jpg"))
plt.figure(figsize=(10,10))
plt.title('Imagem Original')
imagem_carro = cv2.cvtColor(imagem_carro, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_carro)
plt.axis('off')
plt.show()
```

```python
imagem_carro = cv2.imread(str(f"{caminho_imagem}/car.jpg"))
imagem_carro_cinza = cv2.cvtColor(imagem_carro, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.title('Imagem Original')
imagem_carro_cinza = cv2.cvtColor(imagem_carro_cinza, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_carro_cinza)
plt.axis('off')
plt.show()
```

```python
detector_carro = cv2.CascadeClassifier(str(f"{classifier}/cars.xml"))
imagem_carro = cv2.imread(str(f"{caminho_imagem}/car.jpg"))
imagem_carro_cinza = cv2.cvtColor(imagem_carro, cv2.COLOR_BGR2GRAY)
imagem_carro_cinza = cv2.cvtColor(imagem_carro_cinza, cv2.COLOR_BGR2RGB)
deteccao_carros = detector_carro.detectMultiScale(imagem_carro_cinza, scaleFactor=1.04)
for (x, y, w, h) in deteccao_carros:
    cv2.rectangle(imagem_carro, (x, y), (x + w, y + h), (0,255,0), 1)
plt.figure(figsize=(10,10))
plt.title('Imagem Original')
imagem_carro = cv2.cvtColor(imagem_carro, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_carro)
plt.axis('off')
plt.show()
```

## Detecção de Relógios

```python
detector_relogio = cv2.CascadeClassifier(str(f"{classifier}/clocks.xml"))
imagem_relogio = cv2.imread(str(f"{caminho_imagem}/clock.jpg"))
imagem_relogio_cinza = cv2.cvtColor(imagem_relogio, cv2.COLOR_BGR2GRAY)
imagem_relogio_cinza = cv2.cvtColor(imagem_relogio_cinza, cv2.COLOR_BGR2RGB)
deteccao_relogios = detector_carro.detectMultiScale(imagem_relogio_cinza, scaleFactor=1.03)
for (x, y, w, h) in deteccao_relogios:
    cv2.rectangle(imagem_relogio, (x, y), (x + w, y + h), (0,255,0), 1)
plt.figure(figsize=(10,10))
plt.title('Imagem Original')
imagem_relogio = cv2.cvtColor(imagem_relogio, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_relogio)
plt.axis('off')
plt.show()
```

```python
detector_pessoa = cv2.CascadeClassifier(str(f"{classifier}/fullbody.xml"))
imagem_pessoa = cv2.imread(str(f"{caminho_imagem}/people3.jpg"))
# imagem_pessoa = cv2.resize(imagem_pessoa, (900, 600))
imagem_pessoa_cinza = cv2.cvtColor(imagem_pessoa, cv2.COLOR_BGR2GRAY)
imagem_pessoa_cinza = cv2.cvtColor(imagem_pessoa_cinza, cv2.COLOR_BGR2RGB)
deteccao_pessoa = detector_carro.detectMultiScale(imagem_pessoa_cinza, scaleFactor=1.01)
for (x, y, w, h) in deteccao_pessoa:
    cv2.rectangle(imagem_pessoa, (x, y), (x + w, y + h), (0,255,0), 1)
plt.figure(figsize=(10,10))
plt.title('Imagem Original')
imagem_pessoa = cv2.cvtColor(imagem_pessoa, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_pessoa)
plt.axis('off')
plt.show()
```
