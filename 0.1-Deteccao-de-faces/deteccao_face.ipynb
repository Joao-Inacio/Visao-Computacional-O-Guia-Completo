## Carregamento da imagem

```python
import cv2
from pathlib import Path
```

```python
BASE_DIR = Path.cwd().parent

caminho_imagem = BASE_DIR / "data" / "Images"
```

```python
imagem = cv2.imread(str(f"{caminho_imagem}/people1.jpg"))
```

```python
imagem.shape
```

```python
cv2.imshow("img",imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imagem = cv2.resize(imagem, (800, 600))
```

```python
cv2.imshow("img",imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
cv2.imshow("img",imagem_cinza)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
