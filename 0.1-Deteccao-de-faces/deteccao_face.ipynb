```python
BASE_DIR = Path.cwd().parent

caminho_imagem = BASE_DIR / "data" / "Images"
classifier = BASE_DIR / "data" / "Cascades"
```

```python
imagem = cv2.imread(str(f"{caminho_imagem}/people1.jpg"))
```

```python
imagem.shape
```

```python
cv2.imshow("img",imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imagem = cv2.resize(imagem, (800, 600))
```

```python
cv2.imshow("img",imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
cv2.imshow("img",imagem_cinza)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imagem_cinza.shape
```

## Detecção de faces

```python
detector_facial = cv2.CascadeClassifier(str(f"{classifier}/haarcascade_frontalface_default.xml"))
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza)
deteccoes
```

```python
for x, y, w, h in deteccoes:
    cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 100, 105), 1)
cv2.imshow("img",imagem)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
